Pandas is a popular library in Python for data manipulation and analysis. It provides a wide range of functionalities for working with data in a machine learning context. Here are some of the most frequent functionalities that are used in machine learning work:

Data loading and saving: Pandas provides functions such as $\textbf{pd.read_csv()}$, $\textbf{pd.read_excel()}$, $\textbf{pd.read_json()}$ to load data from different file formats, and $\textbf{df.to_csv()}$, $\textbf{df.to_excel()}$, $\textbf{df.to_json()}$ to save data to different file formats.

Data exploration: Pandas provides functionalities such as $\textbf{df.head()}$, $\textbf{df.tail()}$, $\textbf{df.describe()}$, $\textbf{df.info()}$ to explore the data, check for missing values, and get an overview of the data.

Data cleaning: Pandas provides functionalities such as $\textbf{df.dropna()}$, $\textbf{df.fillna()}$, $\textbf{df.replace()}$, $\textbf{df.drop_duplicates()}$ to clean the data, handle missing values, and remove duplicate records.

Data transformation: Pandas provides functionalities such as $\textbf{df.groupby()}, \textbf{df.pivot_table()}$, $\textbf{df.melt()}$, $\textbf{df.unstack()}$ to transform the data, group by certain columns, reshape the data, and create pivot tables.

Data selection and filtering: Pandas provides functionalities such as df[column_name], df.loc[row_indexer,column_indexer], df.iloc[row_indexer,column_indexer] to select and filter the data, slice the dataframe and access specific columns and rows.

Merging and joining: Pandas provides functionalities such as $\textbf{pd.merge()}$, $\textbf{df.join()}$ to merge and join multiple dataframes, combine data from different sources, and join tables on specific columns.

Data visualization: Pandas provides functionalities such as $\textbf{df.plot()}$, $\textbf{df.hist()}$, $\textbf{df.scatter()}$ to visualize the data, plot histograms, scatter plots, and other types of charts using the data.

These are just a few examples of the most frequently used functionalities of pandas in machine learning work, but it offers more functionalities that can be used according to the needs and requirements of the project.

# Create a Test Set

```python
import numpy as np

def shuffle_and_split_data(data, test_ratio):
    shuffled_indices = np.random.permutation(len(data))
    test_set_size = int(len(data) * test_ratio)
    test_indices = shuffled_indices[:test_set_size]
    train_indices = shuffled_indices[test_set_size:]
    return data.iloc[train_indices], data.iloc[test_indices]
```

```python
train_set, test_set = shuffle_and_split_data(housing, 0.2)
len(train_set)
```

# Custom Transformers

```python
housing.columns
```

Let us create custom transformers that will add some attributes

  -- rooms per household
    
  -- population per household

  -- bedrooms per room

```python
housing.head(10)
```

```python
from sklearn.base import BaseEstimator, TransformerMixin
rooms_ix, bedrooms_ix, population_ix, households_ix = 3, 4, 5, 6

class CombinedAttributesAdder(BaseEstimator, TransformerMixin):
    def __init__(self, add_bedrooms_per_room = True):
        self.add_bedrooms_per_room = add_bedrooms_per_room
    def fit(self, X, y = None):
        return self # do nothing
    def transform(self, X):
        rooms_per_household = X[:, rooms_ix]/X[:, households_ix]
        population_per_household = X[:, population_ix] / X[:, households_ix]
        if(self.add_bedrooms_per_room):
            bedrooms_per_room = X[:, bedrooms_ix] / X[:, rooms_ix]
            return np.c_[X, rooms_per_household, population_per_household, bedrooms_per_room]
        else:
            return np.c_[X, rooms_per_household, population_per_household]
```

```python
attr_addr = CombinedAttributesAdder(add_bedrooms_per_room = False)
housing_extra_attributes = attr_addr.transform(housing.values)
housing_extra_attributes
```
